```python
import numpy as np
import healpy as hp
```

```python
fmap = hp.read_map('data/total_healpix_55_Y_z6_R30_T150_m2_AMEGO.fits')
```

```python
hp.mollview(fmap, title="From Guili",norm='hist')
hp.graticule()
```

```python
fmap_smoothed = hp.smoothing(fmap,fwhm=np.radians(1.0))
```

```python
hp.mollview(fmap_smoothed, title="Counts map smeared 1 deg",norm='hist')
hp.graticule()
```
